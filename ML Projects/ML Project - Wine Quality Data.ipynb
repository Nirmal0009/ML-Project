```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif

dfinfogain = df.copy()
dfinfogain['quality'] = qualityCol
dfinfogain = dfinfogain.drop(columns=['type'])

# Split the data into features and labels
X = dfinfogain.drop(columns=['quality'])
y = dfinfogain['quality']

# Calculate the information gain for each feature
info_gain = mutual_info_classif(X, y)
info_gain_df = pd.DataFrame({'feature': X.columns, 'info_gain': info_gain})
info_gain_df = info_gain_df.sort_values('info_gain', ascending=False)
info_gain_df
```

```python
##---------Type the code below this line------------------##
print('Gini Index:')

giniIndexes = []

for col in dfinfogain.columns:
  lst = dfinfogain[col].values.tolist()
  num_vars = pd.Series(lst)
  probs = num_vars.value_counts(normalize=True)
  gini_index = 1 - np.sum(np.square(probs))
  if col != 'quality':
    giniIndexes.append(gini_index)

gini_index_df = pd.DataFrame({'feature': X.columns, 'gini_index': giniIndexes})
gini_index_df = gini_index_df.sort_values('gini_index', ascending=False)
gini_index_df
```

```python
from scipy.stats import entropy

target_entropy = entropy(y.value_counts(normalize=True), base=2)
gain_ratio = []
for i in range(len(info_gain)):
    split_info = entropy(X.iloc[:, i].value_counts(normalize=True), base=2)
    gain_ratio.append(info_gain[i] / split_info)

gain_ratio_df = pd.DataFrame({'feature': X.columns, 'gain_ratio': gain_ratio})
gain_ratio_df = gain_ratio_df.sort_values('gain_ratio', ascending=False)
gain_ratio_df
```

## 6.1 ML technique 1 + Justification

```python
##---------Type the code below this line------------------##
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

dfml = df.copy()
dfml = dfml.drop(columns=['type'])

lab_enc = LabelEncoder()
dfml['quality'] = lab_enc.fit_transform(dfml['quality'])
dfml['quality'].unique()

# Split the data into features and labels
Xml = dfml.drop(columns=['quality'])
yml = dfml['quality']

X_train, X_test, Y_train, Y_test = train_test_split(Xml,yml, test_size = 0.4, random_state = 52)
# Splitting dataset into 60% training set and 40% testing set with
# random state of 52.
# Using standard scaling to get optimized result
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)

#SVM
from sklearn.svm import SVC
svm_clsf = SVC()
svm_clsf.fit(X_train, Y_train)
svm_pred = svm_clsf.predict(X_test)

# Support Vector Machines (SVM) is a popular machine learning technique that can be used for regression and classification problems. 
# It is particularly useful for handling non-linear data and can provide good accuracy even with relatively small datasets.
# SVM can be a good choice because it has several numerical features that are likely to have non-linear relationships with the target variable (quality). 
# SVM can also handle high-dimensional data and is known to be effective at reducing overfitting.
```

## 6.2 ML technique 2 + Justification

```python
##---------Type the code below this line------------------##
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rfcs = RandomForestClassifier(n_estimators = 200)
rfcs.fit(X_train, Y_train)
rfcs_pred = rfcs.predict(X_test)

# Random Forest is an ensemble model that combines multiple decision trees to create a more robust and accurate model.

# High Accuracy: Random Forests have been proven to produce high accuracy compared to single decision trees. 
# By aggregating predictions from multiple decision trees, the model can reduce the risk of overfitting and improve the generalization ability of the model.
# Variable Importance: Random Forests can provide information on feature importance, which can be useful in identifying the most important features in a dataset.
# Easy to Implement: Random Forests are relatively easy to implement, and there are several libraries available in various programming languages, making it easy to get started with the algorithm.
# Versatility: Random Forests can be used for both classification and regression tasks, making it a versatile algorithm that can be applied to a wide range of problems.
```
